````python
#1
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("./data/college_completion.csv")
````

````python
#2
print("Data Dimensions:", df.shape)
print("There are 3798 observations")
print("Column Names:", df.columns)
df.head()
````

````python
#3
cross_tab = pd.crosstab(df['control'], df['level'])
print("\nCross-tabulation of control and level:")
print(cross_tab)
````

Patterns:

Public institutions are more common at the 2-year level (926 institutions) than at the 4-year level (632 institutions). This suggests that community colleges (which are typically public) make up a significant portion of 2-year institutions. Private not-for-profit institutions overwhelmingly consist of 4-year institutions (1,180) compared to only 68 at the 2-year level.
````python
#4
plt.figure(figsize=(15,5))
plt.subplot(1,3,1)
df['grad_100_value'].hist(bins=20, edgecolor='black')
plt.title("Histogram of Graduation Rate")

plt.subplot(1,3,2)
sns.kdeplot(df['grad_100_value'], shade=True)
plt.title("Kernel Density Plot")

plt.subplot(1,3,3)
sns.boxplot(y=df['grad_100_value'])
plt.title("Boxplot")

plt.show()
print("\nStatistical Description of grad_100_value:")
print(df['grad_100_value'].describe())
````

````python
#5
plt.figure(figsize=(12,6))
sns.kdeplot(data=df, x='grad_100_value', hue='control', shade=True)
plt.title("Graduation Rate by Control")
plt.show()

plt.figure(figsize=(12,6))
sns.kdeplot(data=df, x='grad_100_value', hue='level', shade=True)
plt.title("Graduation Rate by Level")
plt.show()
````

Observations:

Graduation Rate by Control:
Public institutions (blue) have the highest density at very low graduation rates (around 0-20%). This suggests that a significant number of public institutions have low graduation rates. Private not-for-profit institutions (orange) show a much wider spread, with many institutions having higher graduation rates compared to public institutions. Private for-profit institutions (green) have a relatively flat distribution, indicating that graduation rates are more evenly spread across a range of values, but they tend to have lower graduation rates compared to private not-for-profit institutions.

Graduation Rate by Level:
2-year institutions (orange) have a peak around lower graduation rates, meaning most 2-year colleges tend to have lower graduation rates. 4-year institutions (blue) show a wider spread with a higher probability of achieving better graduation rates compared to 2-year institutions. The overlap between the two distributions suggests that some 2-year institutions have comparable graduation rates to 4-year institutions, but the general trend indicates lower rates for 2-year colleges.


````python
# Grouped statistical descriptions
print("\nGraduation Rate Statistics by Level:")
print(df.groupby('level')['grad_100_value'].describe())

print("\nGraduation Rate Statistics by Control:")
print(df.groupby('control')['grad_100_value'].describe())
````

Private not-for-profit institutions appear to have the highest graduation rates. 4-year institutions generally have higher graduation rates than 2-year institutions. Public institutions tend to have lower graduation rates, with a strong concentration at the lower end.

````python
#6
df['levelXcontrol'] = df['level'] + ', ' + df['control']
plt.figure(figsize=(12,6))
sns.kdeplot(data=df, x='grad_100_value', hue='levelXcontrol', shade=True)
plt.title("Graduation Rate by Level and Control")
plt.show()
````

4 year private not for profit and 2 year private for profit have the best graduation rates. 

````python
#7
plt.figure(figsize=(12,6))
sns.kdeplot(df['aid_value'], shade=True)
plt.title("Kernel Density Plot of Aid Value")
plt.show()

print("\nAid Value Statistics by Level:")
print(df.groupby('level')['aid_value'].describe())

print("\nAid Value Statistics by Control:")
print(df.groupby('control')['aid_value'].describe())
````

Graph shape explanation:
Most students receive aid in the lower range (~$3,000–$10,000). A small portion of students receives very high aid amounts, contributing to the long tail. Differences in aid policies by institution type influence the distribution.

````python
#8
plt.figure(figsize=(10,6))
sns.scatterplot(data=df, x='aid_value', y='grad_100_value')
plt.title("Scatterplot of Aid vs Graduation Rate")
plt.show()

plt.figure(figsize=(12,6))
sns.lmplot(data=df, x='aid_value', y='grad_100_value', hue='level', fit_reg=True)
plt.title("Aid vs Graduation Rate by Level")
plt.show()

plt.figure(figsize=(12,6))
sns.lmplot(data=df, x='aid_value', y='grad_100_value', hue='control', fit_reg=True)
plt.title("Aid vs Graduation Rate by Control")
plt.show()
````

The first scatterplot (Aid vs. Graduation Rate) shows a positive trend, meaning that as financial aid increases, graduation rates also tend to be higher. However, there is a lot of dispersion in the lower aid values, with many institutions having low graduation rates despite varying levels of aid.

The second scatterplot groups institutions into 2-year (orange) and 4-year (blue) institutions. The pattern reveals that 2-year institutions tend to have lower graduation rates overall, even when aid increases. 4-year institutions show a stronger positive relationship between aid and graduation rates, suggesting that more aid is linked to higher graduation rates in these institutions. This suggests that financial aid has a more noticeable impact on 4-year institutions compared to 2-year institutions.

The third scatterplot groups institutions by Public (blue), Private not-for-profit (orange), and Private for-profit (green) institutions. Observations include that Private not-for-profit institutions show the strongest positive trend—higher aid is strongly associated with higher graduation rates. Public institutions also show a positive trend, but with more variability. Private for-profit institutions have a weaker relationship between aid and graduation rates, with many having low graduation rates despite receiving aid.
This suggests that financial aid is most effective at increasing graduation rates in private not-for-profit institutions, followed by public institutions. Private for-profit institutions do not show the same benefit from financial aid in terms of graduation rates.
